```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#criando df com dados de entrada e saída
def criar_df(tamanho, preco):
  tamanho_casas = []
  preco_casas = []

  for i in range(5):
    tamanho_casas.append(tamanho)
    preco_casas.append(preco)
    tamanho += 500
    preco += 20000

    df = pd.DataFrame(
      {
          "tamanho_casas": tamanho_casas,
          "preco_casas": preco_casas
      }
    )

  return df

df = criar_df(1000, 30000)
```

```python
#média dos dados
mean_x = df['tamanho_casas'].mean()
mean_y = df['preco_casas'].mean()
```

```python
#quantidade de amostras
n = len(df)
```

```python
#calculo manual do coeficiente da regressão linear
numerador = ((df['tamanho_casas'] - mean_x) * (df['preco_casas'] - mean_y)).sum()
denominador = ((df['tamanho_casas'] - mean_x ** 2)).sum()

w = numerador / denominador
b = mean_y - (w * mean_x)
```

```python
#previsões para novos tamanhos de casas
tamanhos_novos = pd.Series([1200, 1700, 2200])
precos_previstos = w * tamanhos_novos + b
```

```python
#imprimir os coeficientes da regressão
print('Coeficiente angular (w): ', w)
print('Coeficiente Linear (b): ', b)
```

```python
#imprimir previsões
for i in range(len(tamanhos_novos)):
  print(f'Para um tamanho de casa de {tamanhos_novos[i].round(2)}, pés quadrados, o preço previsto é: {precos_previstos[i].round(2)}')
```

```python
import matplotlib.pyplot as plt

# Plotar os dados originais
plt.scatter(df['tamanho_casas'], df['preco_casas'], color='blue', label='Dados Originais')

# Plotar a linha de regressão
plt.plot(tamanhos_novos, precos_previstos, color='red', label='Linha de Regressão')

# Plotar as previsões
plt.scatter(tamanhos_novos, precos_previstos, color='green', label='Previsões')

# Configurações do gráfico
plt.xlabel('Tamanho da Casa (pés quadrados)')
plt.ylabel('Preço da Casa')
plt.title('Regressão Linear para Previsão de Preço de Casa')
plt.legend()

# Mostrar o gráfico
plt.show()
```
